```python
# if we want to know which type of exception we are getting
a= input("enter your number 1 (dividend)")
b= input("enter your number 2 (divisor)")
c= input("enter your value to convert in int")
try:
    a=int(a)
    b=int(b)
    print(a/b)
    c=int(c)
except Exception as E: #Exception is parent of all the exceptions
    print("we encountered an error . ",E)
else :
    print("no error encounterd")
finally:
    print("All done")
```

```python
def take_age(age):
    try:
        age_int=int(age)
        if age_int<=0 or age_int>100:
            raise ValueError("Age is not in range,try with a valid age")
        else :
            print("age is valid : ",age_int)
    except Exception as E:
        print(E)
    finally:
        print("code executed successfully")
```

```python
take_age(123)

```

```python
take_age(-1)
```

```python
take_age(18)
```
